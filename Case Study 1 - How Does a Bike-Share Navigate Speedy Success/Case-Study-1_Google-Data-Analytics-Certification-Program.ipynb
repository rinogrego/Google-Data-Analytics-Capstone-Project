It can be seen that the datatype of each of the data is in string. Let's convert them into datetime.datetime data.

```python
import datetime

def transform_datetime(tgl='2021-03-20 14:10:41'):
    year = int(tgl[:4])
    month = int(tgl[5:7])
    date = int(tgl[8:10])
    hour = int(tgl[11:13])
    minute = int(tgl[14:16])
    second = int(tgl[17:19])
    
    return datetime.datetime(year, month, date, hour, minute, second)
```

```python
df_all['started_at'] = df_all['started_at'].apply(lambda x: transform_datetime(x))
df_all['ended_at'] = df_all['ended_at'].apply(lambda x: transform_datetime(x))
```

```python
df_all['time_spent'] = df_all['ended_at'] - df_all['started_at']
df_all['time_spent_seconds'] = (df_all['ended_at'] - df_all['started_at'])/np.timedelta64(1, 's')
df_all.head()
```

Now we can do mathematical calculations with the time_spent data.

```python
all_time_spent_mean = df_all['time_spent'].mean()
all_time_spent_max = df_all['time_spent'].max()
all_time_spent_min = df_all['time_spent'].min()
all_time_spent_std = df_all['time_spent'].std()

print("Casual Rider Statistic")
print("Mean:", all_time_spent_mean)
print("Max:", all_time_spent_max)
print("Min:", all_time_spent_min)
print("Std:", all_time_spent_std)
```

It will be good if we have more clearer data about distance. Now create the column that store the data about it by process it with latitude and longitude.

```python
from geopy.distance import geodesic
```

```python
distance_val_m = []
distance_val_km = []

baris = 0
for row in df_all_clean.values:
    baris += 1
    # lat start, long start, lat end, long end
    # index: 8, 9, 10, 11
    print(f"Processing row: {baris}  ", end='\r')
    
    lat_start = row[8]
    long_start = row[9]
    lat_end = row[10]
    long_end = row[11]
    
    coords_1 = (lat_start, long_start)
    coords_2 = (lat_end, long_end)
    
    jarak_m = geodesic(coords_1, coords_2).m
    jarak_km = geodesic(coords_1, coords_2).km
    
    distance_val_m.append(jarak_m)
    distance_val_km.append(jarak_km)

df_all_clean['distance_in_meter'] = np.array(distance_val_m)
df_all_clean['distance_in_km'] = np.array(distance_val_km)
df_all_clean.head()
```

Giving new columns to separate each time-related columns into 2 parts (year, month, day) and (hour, minute, second)

```python
def get_date_time_specific(tgl='2021-03-20 14:10:41', type=None):
    if type is None or type not in ['year', 'month', 'date', 'hour', 'minute', 'second']:
        print("type given: None or invalid")
        return None
    
    if type == 'year':
        return int(tgl[:4])
    
    if type == 'month':
        return int(tgl[5:7])
    
    if type == 'day':
        return int(tgl[8:10])
    
    if type == 'hour':
        return int(tgl[11:13])
    
    if type == 'minute':
        return int(tgl[14:16])
    
    if type == 'second':
        return int(tgl[17:19])
```

### Further Analysis: Determining the most effective hour to advertise

Now we have found our potential locations, we can further find out the time where we can advertise our services. To do this, we need to look at the distribution of active hour of our potential stations.

Create a column to contain hour data where the casual riders start riding

```python
casual['hour_started_at'] = casual['time_started_at'].apply(lambda x: x.hour)
```

Now let's simultaneously plot the hour activity distributions of each station

```python
potential_stations = ['Clark St & Elm St', 'Wells St & Concord Ln', 'Wells St & Elm St']
hour_bins = np.arange(0, 25)
hour_ticks = np.arange(0, 25, 2)

fig, axs = plt.subplots(1, 3, figsize=(18, 5))

for idx, station in enumerate(potential_stations):
    axs[idx].hist(casual['hour_started_at'].where(casual['start_station_name'] == station), bins=hour_bins)
    axs[idx].set_title(station)
    axs[idx].set_xticks(hour_ticks)

plt.show()
```

We can see that the distributions of hour activitiy for each station is almost the same, where casual riders start riding between 6-8 o'clock and the activity is peaking at around 18-20 o'clock. **The advertising team can start working at 3PM until 11PM if we want to maximize interaction with customers and still go by 8 hours per day work**.
